```python
import numpy as np 
import matplotlib.pyplot as plt
import glob
import cv2

from keras.models import Model, Sequential
from keras.layers import Dense, Flatten, Conv2D, MaxPooling2D
from keras.layers import BatchNormalization
import os
import seaborn as sns
from keras.applications.vgg16 import VGG16
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
print(os.listdir("./PCOS/data"))
```

```python
SIZE = 256
```

```python
train_images = []
train_labels = [] 
```

```python
for directory_path in glob.glob("./PCOS/data/train/*"):
    label = directory_path.split("\\")[-1]
    print(label)
    for img_path in glob.glob(os.path.join(directory_path, "*.jpg"))[:500]:
        print(img_path)
        img = cv2.imread(img_path, cv2.IMREAD_COLOR)       
        img = cv2.resize(img, (SIZE, SIZE))
        img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
        train_images.append(img)
        train_labels.append(label)
```

```python
train_images = np.array(train_images)
train_labels = np.array(train_labels)
```

```python
test_images = []
test_labels = [] 
c=0
for directory_path in glob.glob("./PCOS/data/train/*"):
    label = directory_path.split("\\")[-1]
    print(label)
    for img_path in glob.glob(os.path.join(directory_path, "*.jpg"))[500:]:
        print(img_path)
        img = cv2.imread(img_path, cv2.IMREAD_COLOR)       
        img = cv2.resize(img, (SIZE, SIZE))
        img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
        test_images.append(img)
        test_labels.append(label)
```

```python
test_images = np.array(test_images)
test_labels = np.array(test_labels)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(test_labels)
test_labels_encoded = le.transform(test_labels)
le.fit(train_labels)
train_labels_encoded = le.transform(train_labels)

#Split data into test and train datasets (already split but assigning to meaningful convention)
x_train, y_train, x_test, y_test = train_images, train_labels_encoded, test_images, test_labels_encoded
```

```python
# Normalize pixel values to between 0 and 1
x_train, x_test = x_train / 255.0, x_test / 255.0

#One hot encode y values for neural network. 
# from keras.utils import to_categorical
# y_train_one_hot = to_categorical(y_train)
# y_test_one_hot = to_categorical(y_test)
```

```python
nsamples, nx, ny, nrgb = x_train.shape
x_train2 = x_train.reshape((nsamples,nx*ny*nrgb))
```

```python
nsamples, nx, ny, nrgb = x_test.shape
x_test2 = x_test.reshape((nsamples,nx*ny*nrgb))
```

# Random Forest Regressor

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train2,y_train)
```

```python
y_pred=rf.predict(x_test2)
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,r2_score
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
```

```python
confusion_matrix(y_pred,y_test)
```

## Gradient Boosting 

```python
from sklearn.ensemble import GradientBoostingClassifier
gb=GradientBoostingClassifier(n_estimators=500,learning_rate=0.001,random_state=100,max_features=7)
gb.fit(x_train2,y_train)
```

```python
y_pred=gb.predict(x_test2)
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
```

```python
confusion_matrix(y_pred,y_test)
```

## Logistic Regression

```python
from sklearn import linear_model
lr = linear_model.LogisticRegression()
lr.fit(x_train2,y_train)
```

```python
y_pred=lr.predict(x_test2)
y_pred
```

```python
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
```

```python
confusion_matrix(y_pred,y_test)
```
